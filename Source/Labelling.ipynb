## Tính điểm tổ hợp

```python
import pandas as pd
# Đọc file CSV
df_diemthi = pd.read_csv("diemthi.csv")

df_diemchuan = pd.read_csv("diemchuan_ABCD(2).csv")

df_tohop = pd.read_csv("tohop_mon(2).csv")

```

```python
# Tính điểm tổ hợp
# Hàm tính điểm 1 tổ hợp cho 1 dòng dữ liệu
import pandas as pd
# Đọc file CSV
def tinh_diem_to_hop(row, mon1, mon2, mon3):
    def lay_diem(mon):
        if mon == 'KHTN':
            return np.nanmean([row.get('Lí'), row.get('Hóa'), row.get('Sinh')])
        elif mon == 'KHXH':
            return np.nanmean([row.get('Sử'), row.get('Địa'), row.get('GDCD')])
        else:
            return row.get(mon)

    diem1 = lay_diem(mon1)
    diem2 = lay_diem(mon2)
    diem3 = lay_diem(mon3)

    # Nếu bất kỳ môn nào là NaN thì không tính tổ hợp
    if pd.isna(diem1) or pd.isna(diem2) or pd.isna(diem3):
        return np.nan
    return round(diem1 + diem2 + diem3, 2)

# Tạo DataFrame để lưu điểm các tổ hợp
df_tohop_diem = df_diemthi[['SBD']].copy()
df_tohop_diem['Năm thi'] = df_diemthi['Năm thi']

# Lặp qua từng tổ hợp và tính điểm
for _, row in df_tohop.iterrows():
    ma_tohop = row['Tổ hợp']
    mon1 = row['Môn 1']
    mon2 = row['Môn 2']
    mon3 = row['Môn 3']
    
    df_tohop_diem[ma_tohop] = df_diemthi.apply(lambda r: tinh_diem_to_hop(r, mon1, mon2, mon3), axis=1)

# Kết quả
df_tohop_diem.head()
```

```python
# Dán nhãn các ngành trúng tuyển cho từng thí sinh
import pandas as pd

# Giả sử df_diemchuan và df_tonghop đã được load từ CSV hoặc DataFrame có sẵn

def xet_trung_tuyen(df_diemchuan, df_tonghop):
    # Đảm bảo cột điểm tổ hợp là float
    df_diemchuan["Điểm chuẩn"] = pd.to_numeric(df_diemchuan["Điểm chuẩn"], errors="coerce")
    
    # Lưu kết quả vào danh sách
    ket_qua = []

    # Duyệt qua từng dòng trong df_tonghop
    for _, ts_row in df_tonghop.iterrows():
        sbd = ts_row["SBD"]
        nam_thi = ts_row["Năm thi"]
        diem_nn = ts_row.get("Ngoại ngữ", 0)

        # Lọc điểm chuẩn theo năm
        df_diem_nam = df_diemchuan[df_diemchuan["Năm"] == nam_thi]

        # Duyệt qua từng dòng điểm chuẩn trong năm đó
        for _, dc_row in df_diem_nam.iterrows():
            to_hop = dc_row["Tổ hợp"]
            loai_diem = dc_row["Loại điểm"]
            diem_chuan = dc_row["Điểm chuẩn"]

            # Lấy điểm tổ hợp tương ứng từ df_tonghop
            diem_ts = ts_row.get(to_hop, None)
            if pd.isna(diem_ts):
                continue

            # Nếu là "Thang 40", cộng thêm điểm Ngoại ngữ
            if "Thang 40" in loai_diem:
                diem_ts += diem_nn if not pd.isna(diem_nn) else 0

            # So sánh với điểm chuẩn
            if diem_ts >= diem_chuan:
                ket_qua.append({
                    "SBD": sbd,
                    "Năm": nam_thi,
                    # "Điểm thí sinh": diem_ts,
                    # "Điểm chuẩn": diem_chuan,
                    "Tên ngành trúng tuyển": dc_row["Tên Ngành"],
                    "Tên trường trúng tuyển": dc_row["Tên Trường"]
                })

    # Tạo DataFrame kết quả
    df_ketqua = pd.DataFrame(ket_qua)
    return df_ketqua
```
